```python
def run_grid_search(graph, parameter_space, matching_size, gradual_mutation=False, fitness_method=1):
    graph_res = []
    for params in parameter_space:
        population_size, generations, mutation_rate, elitism_count = params
        best_solution = genetic_algorithm(graph, population_size, generations, mutation_rate, elitism_count, gradual_mutation, fitness_method)
        best_solution_size = sum(best_solution)
        approximation_ratio =  matching_size / best_solution_size
        graph_res.append((params, best_solution_size, approximation_ratio))
        print(f"Parameters: {params}, Solution Size: {best_solution_size}, Approximation Ratio: {approximation_ratio:.2f}")
    return graph_res
```

```python
import matplotlib.pyplot as plt

def prepare_plot_data(results):
    plot_data = []
    for config, res in results.items():
        for params, solution_size, approximation_ratio in res:
            plot_data.append({
                "config": config,
                "params": params,
                "solution_size": solution_size,
                "approximation_ratio": approximation_ratio
            })
    return plot_data

def plot_results(plot_data, matching_size, greedy_solution_size, title):
    # Separate data by configuration
    configs = list(set(d['config'] for d in plot_data))
    
    # Create a plot for each configuration
    plt.figure(figsize=(15, 10))
    
    for i, config in enumerate(configs):
        data = [d for d in plot_data if d['config'] == config]
        
        # Extract parameter values and corresponding solution sizes and approximation ratios
        population_sizes = [d['params'][0] for d in data]
        generation_counts = [d['params'][1] for d in data]
        mutation_rates = [d['params'][2] for d in data]
        elitism_percents = [d['params'][3] for d in data]
        solution_sizes = [d['solution_size'] for d in data]
        approximation_ratios = [d['approximation_ratio'] for d in data]

        # Create scatter plot
        plt.subplot(2, 2, i + 1)
        plt.scatter(range(len(solution_sizes)), solution_sizes, c=approximation_ratios, cmap='viridis', marker='o')
        plt.colorbar(label='Approximation Ratio')
        plt.axhline(y=matching_size, color='r', linestyle='--', label=f'Max Matching Solution Size: {matching_size}')
        plt.axhline(y=greedy_solution_size, color='b', linestyle='--', label=f'Greedy Solution Size: {greedy_solution_size}')
        plt.xlabel('Parameter Configuration Index')
        plt.ylabel('Solution Size')
        plt.title(config)
        plt.legend()
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```
